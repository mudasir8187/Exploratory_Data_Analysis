# Exploratory Data Analysis

- Import Libraries

```python
import pandas as pd
import seaborn as sns 
```

```python
# Loading built-in Data_Set
titanic=sns.load_dataset("titanic")
```

```python
# First Five records
titanic.head()
```

```python
# Info about Data_Set
titanic.info()
```

```python
# Statistical info about Data_set
titanic.describe()
```

# Cleaning and Filtering Data in Data_set

```python
# checking for null values
null_values=titanic.isnull().sum()
null_values
```

- Here deck column has maximum null values so we drop it

```python
clean=titanic.drop(["deck"],axis=1)
clean.shape
```

** Data_set has more null values if we remove the whole columns then it will affect the rows and data so thats why we remove the rows only **

```python
cleaned_data=clean.dropna()
cleaned_data.shape
```

# Difference between Cleaned data and UnCleaned Data

```python
# Checking for cleaned data
cleaned_data.describe()
```

```python
# UnCleaned Data Set
titanic.describe()
```

```python
boxplot=sns.boxplot(x='sex',y="age",data=cleaned_data)
```

# Removing Outliers 

```python
cleaned_data=cleaned_data[cleaned_data['age']<62]
cleaned_data.head()
```

```python
boxplot=sns.boxplot(x='sex',y="age",data=cleaned_data)
```

# Grouping Data

```python
group_data=cleaned_data.groupby(['sex','class','survived']).mean()
group_data.head()
```

# Co-Relation

```python
corr_data=cleaned_data.corr()
corr_data
```

```python
sns.heatmap(corr_data,annot=True)
```

# Relation Plot

```python
sns.relplot(x='age',y='fare',hue="sex",data=cleaned_data)
```

# BarPlot

```python
sns.barplot(x='sex',y='age',hue='who',data=cleaned_data)
```
